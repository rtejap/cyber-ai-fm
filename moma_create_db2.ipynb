```python
import sqlite3
conn = sqlite3.connect("moma2c.db")
c = conn.cursor()
```

```python
import pandas as pd
artists = pd.read_csv('Artists.csv')
print(artists.shape)
artworks = pd.read_csv('Artworks.csv')
print(artworks.shape)
```

```python
artists.head(5)
```

```python
artists.columns = ['const_id','nme_id','bio_id','nat_id','gen_id','dt1_od','dt2_id','wiki_id','ulan_id']
```

```python
artists.head(5)
```

```python
artworks.head(5)
```

```python
artists.to_sql('artists2', conn, if_exists = 'append', index= False)
```

```python
artworks.to_sql('artworks2', conn, if_exists = 'append', index= False)
```
